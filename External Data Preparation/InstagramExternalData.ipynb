### so we don't have any missing values 

#### Proble: We notice that the comments are written in different ways:  French, English and Tunisian dialect 

#### Solution : We need to detect the language of the comment, translate the text from French to English and create a dictionnary for the Tunisian Dialect (NLP)


```python
#language detection 
import spacy
#from spacy_cld import LanguageDetector

from langdetect import detect
#import langdetect
#import langid
```

```python
languages_langdetectO = []

# the try except blook because there is some tweets contain links
for comment in commentO:
    try:
        result = detect(comment)
        if (result != 'fr' and result != 'en'):
            result='tn'
        
    except:
        result = 'tn or emojis'
    
    finally:
        languages_langdetectO.append(result)
```

```python
languages_langdetectO
```

```python
commentsOrange['language'] = languages_langdetectO
commentsOrange
```

```python

french = commentsOrange.loc[commentsOrange['language']=='fr'].count()[0]
english = commentsOrange.loc[commentsOrange['language']=='en'].count()[0]
tunisian = commentsOrange.loc[commentsOrange['language']=='tn'].count()[0]
emojisOrTN=commentsOrange.loc[commentsOrange['language']=='tn or emojis'].count()[0]


labels = ['French', 'English', 'Tunisian Dialect','TN or emojis']
plt.figure(figsize=(10, 10))
plt.pie([french, english, tunisian,emojisOrTN],labels = labels, shadow = True, autopct='%1.1f%%')
plt.title('Distribution of languages in comments')
plt.legend()
plt.show()
```

  #### Problem: 96.3 the comments are in tunisian dialect therefore we need to create a tunisian dialect dictionnary

```python
from googletrans import Translator
```

```python
translator= Translator()
p=[]
comment_list = comments['Comments'].tolist()
print(comment_list)

for translation in comment_list :
    translations = translator.translate(translation, dest='ar',src='auto')
    for translation1 in translations:
        print(translation1.origin, ' -> ', translation1.text)
        p.append(translation)
print(p)
```
