```python
#remplissage colonne type reponse et operateur
j=0
for i in df.columns:
    if df[i].values.dtype == 'float64':
        data['type reponse'][j]= 'note' 
    else:
        data['type reponse'][j]= 'String'    
    data['operator'][j]= 'operateur principale'
    if 'Orange' in i:
        data['operator'][j]= 'Orange'
    if 'orange' in i:
        data['operator'][j]= 'Orange'
    if 'Ooredoo' in i:
        data['operator'][j]= 'Ooredoo'
    if 'ooredoo' in i:
        data['operator'][j]= 'Ooredoo'
    if 'Tunisie Telecom' in i:
        data['operator'][j]= 'Tunisie Telecom'
    j=j+1
    
```

```python
#remplissage colonne service name
j=0
for i in df.columns:
    data['type question'][j]= 'specifique'
    if 'facture' in i:
        data['service name'][j]= 'facturation'
    if 'Facture' in i:
        data['service name'][j]= 'facturation'
    if 'ecommanderiez' in i:
        data['type question'][j]= 'globale'
    if 'GLOBALE' in i:
        data['type question'][j]= 'globale'
    if 'un opérateur qui' in i:
        data['type question'][j]= 'globale'
    if 'un opérateur qui' in i:
        data['service name'][j]= 'relation client'
    if 'Mobicash ' in i:
        data['service name'][j]= 'Mobicash '
    if 'Mdinar' in i:
        data['service name'][j]= 'Mdinar'
    if 'oaming' in i:
        data['service name'][j]= 'roaming'
    j=j+1
data
```
